```python
import csv
# Membaca file 'goals_from_counter_attack.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Goals From Counter Attack.csv', 'r') as goals_file:
    goals_reader = csv.DictReader(goals_file)
    goals_data = list(goals_reader)

# Membaca file 'shots_on_target.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Shots On Target.csv', 'r') as shots_on_target_file:
    shots_on_target_reader = csv.DictReader(shots_on_target_file)
    shots_on_target_data = list(shots_on_target_reader)

# Membaca file 'goals_inside_box.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Goals From Inside Box.csv', 'r') as goals_inside_box_file:
    goals_inside_box_reader = csv.DictReader(goals_inside_box_file)
    goals_inside_box_data = list(goals_inside_box_reader)

# Membaca file 'goals_outside_box.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Goals From Outside Box.csv', 'r') as goals_outside_box_file:
    goals_outside_box_reader = csv.DictReader(goals_outside_box_file)
    goals_outside_box_data = list(goals_outside_box_reader)

# Membaca file 'goals_penalty.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Goals From Penalty.csv', 'r') as goals_penalty_file:
    goals_penalty_reader = csv.DictReader(goals_penalty_file)
    goals_penalty_data = list(goals_penalty_reader)

# Membaca file 'hit_woodwork.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Hit Woodwork.csv', 'r') as hit_woodwork_file:
    hit_woodwork_reader = csv.DictReader(hit_woodwork_file)
    hit_woodwork_data = list(hit_woodwork_reader)

# Membaca file 'offsides.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Offsides.csv', 'r') as offsides_file:
    offsides_reader = csv.DictReader(offsides_file)
    offsides_data = list(offsides_reader)

# Membaca file 'shots.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Shots.csv', 'r') as shots_file:
    shots_reader = csv.DictReader(shots_file)
    shots_data = list(shots_reader)

# Membaca file 'goal_freekick.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Goals From Freekick.csv', 'r') as goal_freekick_file:
    goal_freekick_reader = csv.DictReader(goal_freekick_file)
    goal_freekick_data = list(goal_freekick_reader)

# Membaca file 'goal_header.csv'
with open('D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\Attack\Goals From Header.csv', 'r') as goal_header_file:
    goal_header_reader = csv.DictReader(goal_header_file)
    goal_header_data = list(goal_header_reader)

# Menggabungkan data berdasarkan kolom 'Club'
attack_data = []
for goal in goals_data:
    club = goal['Club']
    attack = {
        'Rank': goal['Rank'],
        'Club': club,
        'Goals': goal['Stat'],
        'Shots on Target': None,
        'Goals Inside Box': None,
        'Goals Outside Box': None,
        'Goals Penalty': None,
        'Hit Woodwork': None,
        'Offsides': None,
        'Shots': None,
        'Goals Free Kick': None,
        'Goal Header': None
    }
    for shots in shots_on_target_data:
        if shots['Club'] == club:
            attack['Shots on Target'] = shots['Stat']
            break
    for inside_box in goals_inside_box_data:
        if inside_box['Club'] == club:
            attack['Goals Inside Box'] = inside_box['Stat']
            break
    for outside_box in goals_outside_box_data:
        if outside_box['Club'] == club:
            attack['Goals Outside Box'] = outside_box['Stat']
            break
    for penalty in goals_penalty_data:
        if penalty['Club'] == club:
            attack['Goals Penalty'] = penalty['Stat']
            break
    for woodwork in hit_woodwork_data:
        if woodwork['Club'] == club:
            attack['Hit Woodwork'] = woodwork['Stat']
            break
    for offsides in offsides_data:
        if offsides['Club'] == club:
            attack['Offsides'] = offsides['Stat']
            break
    for shots in shots_data:
        if shots['Club'] == club:
            attack['Shots'] = shots['Stat']
            break
    for freekick in goal_freekick_data:
        if freekick['Club'] == club:
            attack['Goals Free Kick'] = freekick['Stat']
            break
    for header in goal_header_data:
        if header['Club'] == club:
            attack['Goal Header'] = header['Stat']
            break
    attack_data.append(attack)

# Menulis data ke file 'attack.csv'
with open('attack.csv', 'w', newline='') as attack_file:
    fieldnames = [
        'Rank', 'Club', 'Goals', 'Shots on Target', 'Goals Inside Box',
        'Goals Outside Box', 'Goals Penalty', 'Hit Woodwork', 'Offsides',
        'Shots', 'Goals Free Kick', 'Goal Header'
    ]
    writer = csv.DictWriter(attack_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(attack_data)
```

```python
import csv

red_card_file_path = 'D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\General\Red Cards.csv'
yellow_card_file_path = 'D:\GitHub\Streamlit-PL\Streamlit\PL Club Stats\General\Yellow Card.csv'
general_file_path = 'general.csv'

def read_csv_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(row)
    return data

def write_csv_file(file_path, data, fieldnames):
    with open(file_path, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

red_card_data = read_csv_file(red_card_file_path)
yellow_card_data = read_csv_file(yellow_card_file_path)

general_data = []
for red_card, yellow_card in zip(red_card_data, yellow_card_data):
    general_row = {
        'Rank': red_card['Rank'],
        'Club': red_card['Club'],
        'Red Cards': red_card['Stat'],
        'Yellow Cards': yellow_card['Stat']
    }
    general_data.append(general_row)

fieldnames = ['Rank', 'Club', 'Red Cards', 'Yellow Cards']
write_csv_file(general_file_path, general_data, fieldnames)

print(f"File '{general_file_path}' berhasil dibuat.")
```
